```python
'''
The regression plots in seaborn are primarily intended to add a visual guide 
that helps to emphasize patterns in a dataset during exploratory data analyses.
Regression plots as the name suggests creates a regression line between 2 parameters
and helps to visualize their linear relationships. 
'''
```

```python
# import the library
import seaborn as sns
  
# load the dataset
dataset = sns.load_dataset('tips')
  
# the first five entries of the dataset
dataset.head()
```

```python
#Simple linear plot
```

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = dataset)
```

```python
#Linear plot with additional parameters
```

```python

sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = dataset, 
           hue ='sex', markers =['o', 'v'])
```

```python
#Setting the size and color of the plot
```

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = dataset, hue ='sex', 
           markers =['o', 'v'], scatter_kws ={'s':100}, 
           palette ='plasma')
```

```python
#Displaying multiple plots
```

```python
sns.lmplot(x ='total_bill', y ='tip', data = dataset,col ='sex', row ='time', hue ='smoker')
```

```python
#Size and aspect ratio of the plots
```

```python
sns.lmplot(x ='total_bill', y ='tip', data = dataset, col ='sex', 
           row ='time', hue ='smoker', aspect = 0.6, 
           size = 4, palette ='coolwarm')
```
