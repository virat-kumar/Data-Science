```python
import matplotlib.pyplot as plt
import pylab
```

```python
import numpy as np
```

```python
x = np.linspace(0, 20, 1000)
y = np.sin(x)

pylab.plot(x, y)
```

```python
pylab.plot(x, y)
pylab.xlim(5, 15)
pylab.ylim(-1.2, 1.2)
```

```python
pylab.plot(x, y)

pylab.xlabel('this is x!')
pylab.ylabel('this is y!')
pylab.title('My First Plot')
```

```python
y = np.sin(2 * np.pi * x)
pylab.plot(x, y)
pylab.title(r'$\sin(2 \pi x)$')
```

```python
pylab.plot(x, y, '-r')
pylab.xlim(0,10)
pylab.ylim(-1.2, 1.2)

pylab.xlabel('this is x!')
pylab.ylabel('this is y!')
pylab.title('My First Plot')
```

```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

pylab.plot(x, y1, '-b', label = 'sine')
pylab.plot(x, y2, '-r', label = 'cosine')
pylab.legend(loc = 'upper right')
pylab.ylim(-1.5, 2.0)
```

```python
x1 = np.linspace(0, 10, 20)
y1 = np.sin(x1)

x2 = np.linspace(0, 10, 1000)
y2 = np.sin(x2)
```

```python
pylab.plot(x1, y1, 'bo', label = 'sampled')
pylab.plot(x2, y2, ':k', label = 'continuous')
pylab.legend()

pylab.ylim(-1.5, 2.0)
```
