## skewness

The formula for skewness is:

skewness = (1/n) * sum((xi - x_bar)^3) / s^3  
skewness = (1/n) * Sum((observation - mean) ^3) / standared deviation ^3  



where n is the sample size,  
xi is the ith observation,  
x_bar is the sample mean,  
and s is the sample standard deviation.  
  
  This formula calculates the third standardized moment of a distribution, which is a measure of the asymmetry of the distribution. A positive skewness value indicates that the distribution has a longer right tail, while a negative skewness value indicates that the distribution has a longer left tail. A skewness value of 0 indicates a symmetric distribution.

````python
# Define the range of skewness values to generate samples for
skewness_vals = np.linspace(-5, 5, num=11)

# Generate samples for each skewness value and calculate skewness
samples = []
skewness = []
for val in skewness_vals:
    sample = np.random.normal(loc=0, scale=1, size=10000)
    sample = np.power(sample, val) # apply skewness transformation
    samples.append(sample)
    skewness.append(round(skew(sample), 2))

# Plot histograms for each sample
fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(16,10))
for i in range(len(skewness_vals)):
    ax = axs[i // 4, i % 4]
    ax.hist(samples[i], bins=30, color='blue', alpha=0.5)
    ax.set_title(f"Skewness = {skewness[i]}")
plt.show()
````

````python
# Generate a normally distributed sample with no skewness
sample1 = np.random.normal(loc=0, scale=1, size=10000)

# Generate a sample with positive skewness
sample2 = np.random.gamma(2, size=10000)

# Generate a sample with negative skewness
sample3 = np.random.exponential(scale=1, size=10000)

# Calculate skewness for each sample
skew1 = round(skew(sample1), 2)
skew2 = round(skew(sample2), 2)
skew3 = round(skew(sample3), 2)

# Plot the histograms for each sample
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(20,4))
ax1.hist(sample1, bins=30, color='blue', alpha=0.5)
ax1.set_title(f"Normal Distribution (Skewness={skew1})")
ax2.hist(sample2, bins=30, color='green', alpha=0.5)
ax2.set_title(f"Gamma Distribution (Skewness={skew2})")
ax3.hist(sample3, bins=30, color='red', alpha=0.5)
ax3.set_title(f"Exponential Distribution (Skewness={skew3})")
plt.show()
````

````python
def plot(axs, index, arr_data, arr_data_title = ["Positive Skewness", "Negative Skewness", "No Skewness"], after = "", do_skew = True):
    for i in range(len(arr_data)):
        axs[i, index].hist(arr_data[i], bins=30)
        try : 
            axs[i, index].set_title(f"{arr_data_title[i]} {after} skewness : {round(skew(arr_data[i]), 2)}")
        except:
            axs[i, index].set_title(f"{arr_data_title[i]} {after} skewness : NAN")
    
````

````python


def plot_distributions(data):
    fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(8,12))
    fig.tight_layout(pad=3.0)

    # Normal distribution
    mu, std = stats.norm.fit(data)
    x = np.linspace(min(data), max(data), len(data))
    axes[0].hist(data, bins=25, density=True, alpha=0.6)
    axes[0].plot(x, stats.norm.pdf(x, mu, std), linewidth=2)
    axes[0].set_title("Normal Distribution")

    # Poisson distribution
    lambda_ = np.mean(data + np.abs(min(data)))
    x = np.arange(stats.poisson.ppf(0.01, lambda_), stats.poisson.ppf(0.99, lambda_))
    axes[1].hist(data + np.abs(min(data)), bins=25, density=True, alpha=0.6)
    axes[1].plot(x, stats.poisson.pmf(x, lambda_), 'bo', ms=8, label='poisson pmf')
    axes[1].set_title("Poisson Distribution")

    # Binomial distribution
    p = np.mean(data)
    n = len(data)
    x = np.arange(stats.binom.ppf(0.01, n, p), stats.binom.ppf(0.99, n, p))
    axes[2].hist(data, bins=25, density=True, alpha=0.6)
    axes[2].plot(x, stats.binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
    axes[2].set_title("Binomial Distribution")

    # Gamma distribution
    alpha, loc, beta = stats.gamma.fit(data)
    x = np.linspace(min(data), max(data), len(data))
    axes[3].hist(data, bins=25, density=True, alpha=0.6)
    axes[3].plot(x, stats.gamma.pdf(x, alpha, loc, beta), linewidth=2)
    axes[3].set_title("Gamma Distribution")

    # Beta distribution
    alpha, beta, loc, scale = stats.beta.fit(data)
    x = np.linspace(0, 1, len(data))
    axes[4].hist(data, bins=25, density=True, alpha=0.6)
    axes[4].plot(x, stats.beta.pdf(x, alpha, beta, loc, scale), linewidth=2)
    axes[4].set_title("Beta Distribution")

    plt.show()
    
data = np.random.normal(0, 1, size=1000)  # generate random data
plot_distributions(data)
````

#### what is the importance to fix skewness for data scientist?

 
- Fixing skewness is an important step for a data scientist because it can improve the accuracy of statistical models and make the data more interpretable. Skewness refers to the degree of asymmetry in a distribution of data. If a dataset is heavily skewed, it can cause problems with statistical inference, such as underestimating or overestimating the true values of parameters or failing to meet the assumptions of certain statistical tests.

Some of the reasons why it is important to fix skewness in data include: 
1. **Accurate modeling:** Many statistical models assume that the data is normally distributed, and skewed data can violate this assumption. If the data is not normally distributed, modeling techniques like regression and hypothesis testing may not work well, leading to inaccurate results and predictions. 
2. **Improved interpretation:** Skewed data can make it difficult to interpret the results of an analysis. By fixing the skewness, the data can be transformed to a more symmetric distribution, which can help to make the results more interpretable. 
3. **Better decision making:** Accurate data analysis is crucial for making informed decisions. By fixing the skewness, data scientists can ensure that their analysis is based on accurate and reliable data, leading to better decision making.

Overall, fixing skewness is an important step in the data analysis process that can help to ensure accurate modeling, improve the interpretation of results, and lead to better decision making.  
#### How Skeweness effect inter-variable relationships?

 
- Fixing skewness can have an impact on inter-variable relationships in a dataset. Skewness is a measure of the asymmetry in the distribution of a variable, and highly skewed variables can affect the relationships between variables in a dataset. By transforming skewed variables, it may be possible to improve the relationships between variables and better understand how they interact with one another.

Here are some ways in which fixing skewness can affect inter-variable relationships: 
1. Reduce the impact of outliers: Skewed variables are often `associated with outliers`, which can have a significant impact on the relationships between variables. By transforming the skewed variable, it may be possible to reduce the impact of outliers and better understand the relationship between the variables. 
2. Improve normality: If a variable is highly skewed, it may not follow a normal distribution. By transforming the variable to a more normal distribution, it may be possible to better understand the relationship between that variable and other variables in the dataset. 
3. Enhance the strength of correlation: In some cases, transforming a skewed variable can `enhance the strength of the correlation` between variables. This is because skewness can dampen the strength of the relationship between variables, and transforming the variable can help to uncover a stronger relationship.

Overall, fixing skewness can help to improve the inter-variable relationships in a dataset by reducing the impact of outliers, improving normality, and enhancing the strength of correlation. By doing so, data scientists can gain a better understanding of how variables interact with one another and make more accurate predictions and decisions based on the data.


# Statistical Testing
- Statistical testing, also known as hypothesis testing, is a fundamental concept in statistics and data analysis. It is a process of `making decisions or drawing conclusions about a population based on a sample of data.` The aim of statistical testing is to determine whether a hypothesis about a population parameter is supported by the sample data or not.

- In statistical testing, the first step is to formulate a `null hypothesis`, which is a statement that assumes that there is no - significant difference or relationship between the variables being tested. The alternative hypothesis, on the other hand, is a statement that assumes that there is a `significant difference or relationship between the variables`.

- After formulating the null and alternative hypotheses, a statistical test is performed on the sample data. The statistical test involves calculating a `test statistic, which is a measure of how far the observed sample data deviates from what is expected under the null hypothesis.` The test statistic is then compared to a __`critical value`__, which is based on the significance level of the test and the degrees of freedom of the data.

- If the test statistic is `greater than the critical value`, the null hypothesis is rejected, and it is concluded that the alternative hypothesis is supported by the data. If the `test statistic is less than or equal to the critical value`, the null hypothesis is not rejected, and it is concluded that there is not enough evidence to support the alternative hypothesis.

- Statistical testing is used in many areas of data analysis, including in `comparing groups, testing for relationships between variables, and making predictions based on sample data.` It is a powerful tool for making decisions based on data and can help to ensure that conclusions drawn from data are based on sound statistical principles.

#### Common Stastical Tests
1. __T-Test:__ Used to compare the means of two groups and determine if there is a significant difference between them. 
2. __ANOVA (Analysis of Variance)__: Used to compare the means of three or more groups and determine if there is a significant difference between them. 
3. __Chi-Square Test__: Used to test the independence of two categorical variables and determine if they are related to each other. 
4. __Regression Analysis__: Used to model the relationship between two or more variables and determine if there is a significant linear relationship between them. 
5. __Correlation Analysis__: Used to measure the strength and direction of the relationship between two continuous variables. 
6. __Mann-Whitney U Test__: Used to compare the median scores of two independent groups. 
7. __Kruskal-Wallis Test__: Used to compare the median scores of three or more independent groups. 
8. __Wilcoxon Signed-Rank Test__: Used to compare the median scores of two related groups. 
9. __Friedman Test__: Used to compare the median scores of three or more related groups.

#### P-value
- The p-value is a statistical measure that indicates the `probability of observing a test statistic as extreme as, or more extreme than, the observed test statistic, assuming that the null hypothesis is true.` In other words, it is a __measure of the strength of evidence against the null hypothesis__. A smaller p-value indicates stronger evidence against the null hypothesis and in favor of the alternative hypothesis.
- EX:

- Suppose a pharmaceutical company is testing a new drug to lower blood pressure. They conduct a randomized controlled trial with two groups: a treatment group that receives the drug and a control group that receives a placebo. After the trial, they compare the mean blood pressure of the two groups and obtain a p-value of 0.02. This means that if the null hypothesis (that there is no difference in mean blood pressure between the two groups) is true, there is only a 2% chance of observing a difference as large as or larger than the observed difference in mean blood pressure.

- Since the p-value is less than the common significance level of `0.05`, the pharmaceutical company can reject the null hypothesis and conclude that the drug is effective in lowering blood pressure. This conclusion is based on the evidence provided by the p-value, which suggests that the observed difference in mean blood pressure is unlikely to be due to chance alone.

#### Change of critical p-Value
- The critical value can be `changed according to the problem statement for selecting` or rejecting the null hypothesis. The critical value is the value that separates the rejection region from the acceptance region in a hypothesis test. It is based on the chosen level of significance, which is the probability of rejecting the null hypothesis when it is actually true.

The level of significance is typically set at `0.05 or 0.01`, but it can be adjusted based on the problem statement and the goals of the analysis. For example, __`if the consequences of a Type I error (rejecting the null hypothesis when it is actually true) are severe, a lower level of significance may be chosen to reduce the probability of making such an error.`__

The critical value is determined by the probability distribution of the test statistic and the chosen level of significance. For example, if the test statistic follows a `t-distribution` with n degrees of freedom and the level of significance is 0.05, the critical value will be the value of t that separates the lower 2.5% from the upper 97.5% of the distribution (assuming a two-tailed test).

In summary, the critical value can be changed based on the problem statement and the goals of the analysis, but it should be chosen based on sound statistical principles and not arbitrarily.

#### Type | vs Type || error

Type I Error Example:

Suppose a researcher is testing a new drug and wants to see if it is effective in reducing the symptoms of a disease. They conduct a study with a control group and a treatment group, and they use a significance level of 0.05 to test the null hypothesis that there is no difference between the two groups.

If the researcher `incorrectly rejects the null hypothesis` (i.e., makes a Type I error), they conclude that the drug is effective when it is actually not effective. This could lead to the drug being approved for use, even though it does not work, which could harm patients or waste resources.

Type II Error Example:

Suppose a company is testing a new product and wants to see if it is preferred by consumers compared to the old product. They conduct a study with a sample of consumers and use a significance level of 0.05 to test the null hypothesis that there is no difference in preference between the two products.

If the researcher `incorrectly fails to reject the null hypothesis` (i.e., makes a Type II error), they conclude that there is no difference between the two products when there actually is a difference. This could lead the company to continue selling the old product, even though the new product is actually preferred by consumers, resulting in lost sales and profits.

In summary, Type I error occurs when the `null hypothesis is incorrectly rejected, leading to a false positive result`. Type II error occurs when the `null hypothesis is incorrectly accepted, leading to a false negative result`. Both types of errors can have serious consequences and must be carefully considered in statistical hypothesis testing.

### T-test
- The t-test is a statistical hypothesis test used to `compare the means of two groups` and determine `if there is a significant difference between them`. It is typically used when the `sample size is small (less than 30)` and the `population standard deviation is __unknown__`. The test calculates a t-value, which is then compared to a critical value from a t-distribution to determine if the difference between the means is significant.

- __Assumptions__
1. *Normality:* The data should be normally distributed. If the sample size is large enough (typically greater than 30), the central limit theorem may allow for the assumption of normality to be met. 
2. *Independence:* The observations within each group should be independent of each other. In paired samples t-test, the differences between paired observations are assumed to be independent. 
3. *Homogeneity of variance:* The variance within each group should be roughly equal. This assumption is important because if the variances are significantly different between the groups, the t-test may not be appropriate. 
4. *Scale:* The data should be measured on a continuous or ordinal scale.


There are two types of t-tests: 
- `Independent samples t-test` and `paired samples t-test`.  
The independent samples t-test `compares the means of two independent groups`, while the paired samples t-test `compares the means of two related groups`.

Here is an example of how to perform an `independent samples t-test`

Suppose we have two groups of students and we want to test if there is a significant difference between their mean test scores. We have the following data:

Group A: [82, 76, 88, 92, 78]
Group B: [68, 72, 63, 77, 73]



````python
# Independent T-test
group_a = [82, 76, 88, 92, 78]
group_b = [68, 72, 63, 77, 73]

# calculate the t-value and p-value
t_stat, p_val = stats.ttest_ind(group_a, group_b)

# print the results
print("t-value:", t_stat)
print("p-value:", p_val)
````

The paired samples t-test is a statistical hypothesis test used to compare the means of `two related groups and determine if there is a significant difference between them`. It is typically used `when the same group of subjects is measured twice`, such as `before and after an intervention`, and we want to determine if there is a` significant change in the measured variable`. The test calculates a t-value, which is then compared to a critical value from a `t-distribution to determine if the difference between the means is significant.`

Here is an example of how to perform a `paired samples t-test` 

Suppose we have a group of `10 students` who took a math test `before and after a tutoring program`. We want to test if there is a significant difference between their mean test scores before and after the program. We have the following data:

Before: [78, 72, 84, 68, 70, 75, 80, 73, 77, 69]
After: [85, 79, 89, 78, 80, 85, 90, 82, 87, 81]

````python
before = [78, 72, 84, 68, 70, 75, 80, 73, 77, 69]
after = [85, 79, 89, 78, 80, 85, 90, 82, 87, 81]

# calculate the t-value and p-value
t_stat, p_val = stats.ttest_rel(before, after)

# print the results
print("t-value:", t_stat)
print("p-value:", p_val)
````

### ANOVA(Analysis of Variance) 

-  Statistical hypothesis test used to determine if there is a significant difference between the `means of three or more groups`. It compares the variance within each group to the variance between the groups, and calculates an `F-statistic`, which is then compared to a critical value to determine if there is a significant difference.  

- *Assumptions*
1. *Independence:* The observations within each group should be `independent of each other.` 
2. __Normality:__ The data should be normally distributed within each group. If the sample size is large enough, the central limit theorem may allow for the assumption of normality to be met. 
3. __Homogeneity of variance:__ The variance within each group `should be roughly equal`. This assumption is important because if the `variances are significantly different between the groups, the ANOVA test may not be appropriate.` 
4. __Random sampling:__ The samples should be selected randomly from the population.


Example scenario where we can use ANOVA:

Suppose we are studying the effect of three different fertilizers (A, B, and C) on the growth of tomato plants. We have randomly assigned `15 tomato plants to each fertilizer treatment, and after 6 weeks`, we measure the height of each plant. We want to test if there is a significant difference between the mean heights of the plants in each fertilizer group.

Here is the data we have:

Fertilizer A: [20, 21, 24, 19, 22, 25, 20, 23, 26, 21, 22, 20, 23, 24, 19]
Fertilizer B: [25, 27, 24, 26, 29, 28, 27, 25, 24, 26, 23, 28, 29, 27, 25]
Fertilizer C: [30, 31, 29, 33, 32, 28, 30, 33, 31, 29, 32, 30, 29, 34, 31]

````python
# define the three groups
fertilizer_a = [20, 21, 24, 19, 22, 25, 20, 23, 26, 21, 22, 20, 23, 24, 19]
fertilizer_b = [25, 27, 24, 26, 29, 28, 27, 25, 24, 26, 23, 28, 29, 27, 25]
fertilizer_c = [30, 31, 29, 33, 32, 28, 30, 33, 31, 29, 32, 30, 29, 34, 31]

# perform the ANOVA test
f_stat, p_val = stats.f_oneway(fertilizer_a, fertilizer_b, fertilizer_c)

# print the results
print("F-statistic:", f_stat)
print("p-value:", p_val)
````

### Chi-square test 
- Statistical test used to examine the `relationship between two categorical variables`. It is used to determine whether there is a significant association between the two variables or not.

Assumptions of Chi-square test:
- The data should be categorical in nature.
- The sample size should be large enough to ensure that expected cell frequencies are at least 5.

Example scenarios: 
1. An experiment is conducted to determine whether `gender and job satisfaction are associated.` The survey participants are asked to identify their gender and their level of `job satisfaction (satisfied, neutral, dissatisfied)`. The results are summarized in a contingency table. 
2. A company wants to determine if there is an association between `product satisfaction and product rating`. They collected feedback from customers about their product satisfaction and rating on a scale of 1-5.

In both scenarios, the chi-square test can be used to test the association between the two categorical variables.

The p-value in below example is greater than 0.05, indicating that there is no significant association between gender and job satisfaction.

````python
from scipy.stats import chi2_contingency

# Create a contingency table
data = {'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male'],
        'Job Satisfaction': ['Satisfied', 'Neutral', 'Dissatisfied', 'Satisfied', 'Neutral', 'Neutral', 'Dissatisfied', 'Satisfied']}
df = pd.DataFrame(data)

# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Job Satisfaction'])
print(contingency_table)
# Perform chi-square test
stat, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print('Chi-Square Test')
print('Stat:', stat)
print('P-value:', p)
print('Degrees of Freedom:', dof)
print('Expected:', expected)
````

### Regression analysis
- Statistical method used to investigate the `relationship between a dependent variable and one or more independent variables`. It is used to predict the value of the `dependent variable based on the values of one or more independent variables.` Regression analysis can be used for both simple and multiple linear regression.

Assumptions of Regression analysis:
1. *Linearity:* The relationship between the dependent variable and independent variable(s) is linear.
2. *Homoscedasticity:* The variance of the residuals is constant across all levels of the independent variable(s).
3. *Independence:* The observations are independent of each other.
4. *Normality:* The residuals are normally distributed.

Example scenario:
A car dealership wants to predict the `selling price of a car based on its age, mileage, and type (SUV, sedan, or hatchback)`. They collect data on the age, mileage, and type of cars sold in the last year, along with the selling price. They want to use regression analysis to predict the selling price of a car based on its age, mileage, and type.  
Python code for Regression Analysis:

```python

import pandas as pd
import statsmodels.api as sm

# Load the data
data = pd.read_csv('car_data.csv')

# Define the independent variables
X = data[['Age', 'Mileage', 'Type']]

# Convert the 'Type' column to dummy variables
X = pd.get_dummies(X, drop_first=True)

# Define the dependent variable
y = data['Price']

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the model using Ordinary Least Squares (OLS)
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```
